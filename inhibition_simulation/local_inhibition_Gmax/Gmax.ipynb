```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# import excel file
df = pd.read_excel('Gmax.xlsx', sheet_name='Sheet1')

colors = ["#66CDAA","#4682B4","#AB63FA","#FFA15A"]
sns.violinplot(x="group", y="Gmax", data=df, palette=colors, inner="points")
sns.pointplot(x="group", y="Gmax", data=df, markers="o", color="red", errorbar=None, linestyles="none",label=None)
plt.title('Gmax')
```

```python
sns.violinplot(x="group", y="Gmax-Gc", data=df, palette=colors, inner="points")
sns.pointplot(x="group", y="Gmax-Gc", data=df, markers="o", color="red", errorbar=None, linestyles="none",label=None)
plt.title('Gmax-Gc')
```
